**THE BEST DATASET WITH DNN**

```python
import pandas as pd
import numpy as np
from ucimlrepo import fetch_ucirepo
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.inspection import permutation_importance
```

```python
student_performance = fetch_ucirepo(id=320)
X = student_performance.data.features
y = student_performance.data.targets['G3']

# --- انتخاب فیچرهای کلیدی
selected_features = ['school','internet','studytime','famrel','freetime',
                     'goout','Dalc','Walc','absences']
X = X[selected_features]
```

```python
# --- تبدیل categorical ها به عددی
for col in X.columns:
    if X[col].dtype == 'object':
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col])

# --- نرمال‌سازی داده
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- تقسیم داده
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
model = keras.Sequential([
    layers.Dense(128, activation="relu", input_shape=(X_train.shape[1],)),
    layers.Dense(64, activation="relu"),
    layers.Dense(32, activation="relu"),
    layers.Dense(1)
])

model.compile(optimizer="adam", loss="mse", metrics=["mse"])
```

```python
history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=20,
    batch_size=32,
    verbose=0
)
```

```python
y_pred = model.predict(X_test).flatten()

print("MSE:", mean_squared_error(y_test, y_pred))
print("R2:", r2_score(y_test, y_pred))
```

```python
# Permutation Importance
from sklearn.neural_network import MLPRegressor

mlp = MLPRegressor(hidden_layer_sizes=(128,64,32), max_iter=500, random_state=42)
mlp.fit(X_train, y_train)

result = permutation_importance(mlp, X_test, y_test, n_repeats=10, random_state=42)

importances = pd.Series(result.importances_mean, index=selected_features)
importances = importances.sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=importances.values, y=importances.index, palette="viridis")
plt.title("Feature Importance (Permutation Importance)")
plt.xlabel("Mean Importance")
plt.ylabel("Feature")
plt.show()
```

```python
import pandas as pd
import os
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
df = df.drop(columns=["StudentID", "Name", "Study Hours", "Attendance (%)"])
```

```python
imputer_num = SimpleImputer(strategy="mean")
df[["AttendanceRate","StudyHoursPerWeek","PreviousGrade","ExtracurricularActivities","FinalGrade"]] = \
    imputer_num.fit_transform(df[["AttendanceRate","StudyHoursPerWeek","PreviousGrade","ExtracurricularActivities","FinalGrade"]])
```

```python
imputer_cat = SimpleImputer(strategy="most_frequent")
df[["Gender","ParentalSupport","Online Classes Taken"]] = imputer_cat.fit_transform(df[["Gender","ParentalSupport","Online Classes Taken"]])
```

```python
le = LabelEncoder()
for col in ["Gender","ParentalSupport","Online Classes Taken"]:
    df[col] = le.fit_transform(df[col])
```

```python
X = df.drop(columns=["FinalGrade"])
y = df["FinalGrade"]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dense(1, activation="linear")   # خروجی برای رگرسیون
])
```

```python
model.compile(optimizer="adam", loss="mse", metrics=["mae"])
```

```python
history = model.fit(X_train, y_train, validation_split=0.2, epochs=50, batch_size=16, verbose=1)
```

```python
y_pred = model.predict(X_test).flatten()
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("MSE:", mse)
print("R2:", r2)
```
